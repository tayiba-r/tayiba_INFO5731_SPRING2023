Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2012-2022).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# You code here (Please add comments in the code):
import pandas as pd
# using semantic library 
from semanticscholar import SemanticScholar
# calling the semantic scholar object
scholar_object = SemanticScholar()
resultsFromSemantic = []

# using keyword to search the papers within range: The given keyword : information retrieval
# fetching paginated results and store it
paginated_result = scholar_object.search_paper('information retrieval',year="2012-2023")


# Extracting the authors names.
def authorFormat(authors):
  author_names_list = []
  for i in authors:
    author_names_list.append(i.name)
  
  author_names = ", ".join(author_names_list)
  return author_names
title_list=[]
venue_list=[]
year_list=[]
author_list=[]
abstract_list=[]

for data in paginated_result:
  title_data=data.title
  venue_data=data.venue
  year_data=data.year
  author_data=authorFormat(data.authors)
  abstract_data=data.abstract

  title_list.append(title_data)
  venue_list.append(venue_data)
  year_list.append(year_data)
  author_list.append(author_data)
  abstract_list.append(abstract_data)

df1 = pd.DataFrame({'Title':title_data, 'Venue':venue_data, 'Year': year_data, 'Author':author_list, 'Abstract':abstract_data})
df1
```

```python
# You code here (Please add comments in the code):
import snscrape.modules.twitter as sntwitter
import pandas as pd

query = "(from:rihanna)"
tweets = []
limit = 1000


for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    

    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.username, tweet.content])
        
df = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet'])
df
```
